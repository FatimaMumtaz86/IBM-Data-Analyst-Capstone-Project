### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
pip install seaborn
```

```python
import seaborn as sns
```

```python
import random
def map_age_random(age_str):
    try:
        if '-' in age_str:
            age_str = age_str.replace(' years old', '').strip()
            low, high = age_str.split('-')
            return random.randint(int(low.strip()), int(high.strip()))
        elif 'Under' in age_str:
            return random.randint(10, 17)
        elif '65 or older' in age_str:
            return random.randint(65, 75)
    except:
        return None
```

```python
query = """
SELECT Age, ConvertedCompYearly
FROM main
WHERE Age IS NOT NULL AND ConvertedCompYearly IS NOT NULL
"""
df = pd.read_sql_query(query, conn)



random.seed(42)
df['AgeMapped'] = df['Age'].apply(map_age_random)


df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')
df_clean = df.dropna(subset=['ConvertedCompYearly', 'AgeMapped'])

median_salary_by_age = df_clean.groupby('AgeMapped')['ConvertedCompYearly'].median().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=median_salary_by_age, x='AgeMapped', y='ConvertedCompYearly', marker='o')
plt.title('Median Yearly Compensation by Age')
plt.xlabel('Age')
plt.ylabel('Median ConvertedCompYearly (USD)')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python

query = """
SELECT Age, ConvertedCompYearly
FROM main
WHERE Age IS NOT NULL AND ConvertedCompYearly IS NOT NULL
"""
df = pd.read_sql_query(query, conn)


random.seed(42)
df['AgeMapped'] = df['Age'].apply(map_age_random)

df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')
df_filtered = df.dropna(subset=['ConvertedCompYearly', 'AgeMapped'])
df_filtered = df_filtered[(df_filtered['AgeMapped'] >= 25) & (df_filtered['AgeMapped'] <= 45)]

median_by_age = df_filtered.groupby('AgeMapped')['ConvertedCompYearly'].median().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=median_by_age, x='AgeMapped', y='ConvertedCompYearly', marker='o', color='teal')
plt.title('Median Yearly Compensation for Ages 25 to 45')
plt.xlabel('Age')
plt.ylabel('Median ConvertedCompYearly (USD)')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python

query = """
SELECT YearsCodePro, JobSatPoints_6
FROM main
WHERE YearsCodePro IS NOT NULL AND JobSatPoints_6 IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

df_clean = df.dropna(subset=['YearsCodePro', 'JobSatPoints_6'])

satisfaction_by_exp = df_clean.groupby('YearsCodePro')['JobSatPoints_6'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=satisfaction_by_exp, x='YearsCodePro', y='JobSatPoints_6', marker='o', color='purple')
plt.title('Job Satisfaction (JobSatPoints_6) by Years of Experience')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Average Job Satisfaction Score')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python

query = """
SELECT ConvertedCompYearly, WorkExp
FROM main
WHERE ConvertedCompYearly IS NOT NULL AND WorkExp IS NOT NULL
"""
df = pd.read_sql_query(query, conn)
```

```python

df['workExp'] = pd.to_numeric(df['WorkExp'], errors='coerce')
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

df_clean = df.dropna(subset=['WorkExp', 'ConvertedCompYearly'])

median_comp_by_exp = df_clean.groupby('WorkExp')['ConvertedCompYearly'].median().reset_index()

median_comp_by_exp = median_comp_by_exp.sort_values('WorkExp')

plt.figure(figsize=(12, 6))
sns.lineplot(data=median_comp_by_exp, x='WorkExp', y='ConvertedCompYearly', marker='o')
plt.title('Median Compensation Over Work Experience')
plt.xlabel('Years of Work Experience')
plt.ylabel('Median Yearly Compensation (USD)')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python

query = """
SELECT JobSatPoints_6, WorkExp
FROM main
WHERE JobSatPoints_6 IS NOT NULL AND WorkExp IS NOT NULL
"""
df = pd.read_sql_query(query, conn)
```

```python

df['WorkExp'] = pd.to_numeric(df['WorkExp'], errors='coerce')
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

df_clean = df.dropna(subset=['WorkExp', 'JobSatPoints_6'])

mean_sat_by_exp = df_clean.groupby('WorkExp')['JobSatPoints_6'].mean().reset_index()

mean_sat_by_exp = mean_sat_by_exp.sort_values('WorkExp')

plt.figure(figsize=(12, 6))
sns.lineplot(data=mean_sat_by_exp, x='WorkExp', y='JobSatPoints_6', marker='o')
plt.title('Job Satisfaction (JobSatPoints_6) Across Experience Levels')
plt.xlabel('Years of Work Experience')
plt.ylabel('Average Job Satisfaction Score')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
